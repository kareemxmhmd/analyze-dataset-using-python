# Superstore Dataset Analysis

- Load and inspect data
- Clean missing values
- Summary statistics
- Visualizations

## 1. Imports and Configuration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8" if "seaborn-v0_8" in plt.style.available else "ggplot")
sns.set_palette("husl")
```

## 2. Load Data

```python
df = pd.read_excel("Sample - Superstore.xlsx")
df.columns = df.columns.str.strip()

df.head() 
```

## 3. Clean Missing Values

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({"Missing Count": missing, "Missing %": missing_pct})
missing_df
```

## 4. Summary Statistics

```python
df.select_dtypes(include=["number"]).describe().round(2)
```

```python
for col in df.select_dtypes(include=["object", "string"]).columns:
    print(f"\n{col}:")
    display(df[col].value_counts().head(10))
```

## 5. Visualizations

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

df['YearMonth'] = df['Order Date'].dt.to_period('M')
monthly = df.groupby('YearMonth').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
monthly['YearMonth'] = monthly['YearMonth'].astype(str)
```

### Monthly sales and profit (dual-axis line)

```python
fig, ax1 = plt.subplots(figsize=(12, 5))
ax2 = ax1.twinx()

ax1.plot(monthly['YearMonth'], monthly['Sales'], color='steelblue', marker='o', markersize=4, label='Sales')
ax2.plot(monthly['YearMonth'], monthly['Profit'], color='darkorange', marker='s', markersize=4, label='Profit')

ax1.set_xlabel('Month')
ax1.set_ylabel('Sales', color='steelblue')
ax2.set_ylabel('Profit', color='darkorange')
ax1.set_title('Monthly Sales and Profit Over Time — Trend and seasonality', fontsize=12)
ax1.tick_params(axis='x', rotation=45)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

### Category comparison (stacked or grouped bars)

```python
cat_metrics = df.groupby('Category').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
cat_metrics = cat_metrics.sort_values('Sales', ascending=True)

fig, ax = plt.subplots(figsize=(10, 5))
x = np.arange(len(cat_metrics))
width = 0.35
ax.barh(x - width/2, cat_metrics['Sales'], width, label='Sales', color='steelblue')
ax.barh(x + width/2, cat_metrics['Profit'], width, label='Profit', color='darkorange')
ax.set_yticks(x)
ax.set_yticklabels(cat_metrics['Category'])
ax.set_xlabel('Amount')
ax.set_title('Sales and Profit by Category — Revenue vs profitability', fontsize=12)
ax.legend()
plt.tight_layout()
plt.show()
```

### Segment and region (heatmap)

```python
segment_region = df.pivot_table(values='Profit', index='Segment', columns='Region', aggfunc='sum')
fig, ax = plt.subplots(figsize=(8, 5))
sns.heatmap(segment_region, annot=True, fmt='.0f', cmap='RdYlGn', center=0, ax=ax)
ax.set_title('Profit by Segment and Region — Best and worst combinations', fontsize=12)
plt.tight_layout()
plt.show()
```

### Top profitable and loss-making products

```python
prod_profit = df.groupby('Product Name')['Profit'].sum().sort_values()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
prod_profit.tail(10).plot(kind='barh', ax=axes[0], color='green', alpha=0.7)
prod_profit.head(10).plot(kind='barh', ax=axes[1], color='red', alpha=0.7)
axes[0].set_title('Top 10 Profitable Products')
axes[1].set_title('Top 10 Loss-Making Products')
for ax in axes:
    ax.set_xlabel('Profit')
plt.tight_layout()
plt.show()
```

### Sales by sub-category (horizontal bar)

```python
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=True)
fig, ax = plt.subplots(figsize=(10, 8))
subcat_sales.plot(kind='barh', ax=ax, color=plt.cm.viridis(np.linspace(0.2, 0.8, len(subcat_sales))))
ax.set_title('Sales by Sub-Category — Granular product performance', fontsize=12)
ax.set_xlabel('Total Sales')
plt.tight_layout()
plt.show()
```
